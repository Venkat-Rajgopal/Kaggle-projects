```python
# import all libraries
import numpy as np
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Dropout, MaxPooling2D, Flatten, Reshape
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

# drop id column
train = train.drop(['id'], axis=1)
test = test.drop(['id'], axis=1)

train.head()
```

```python
# save targets 
y = train['target']
binarized_labels = label_binarize(y, classes=np.arange(9))
binarized_labels.shape
```

```python
train = train.drop(['target'], axis=1)
```

```python
print('train size:', train.shape)
print('test size:', test.shape)
```

```python
# Split training samples for train and val. 
X_train, X_val, y_train, y_val = train_test_split(train, binarized_labels, test_size = 0.2)
print('Train size:', X_train.shape)
print('Val size:', X_val.shape)
print('Train labels:', y_train.shape)
print('Val labels:', y_val.shape)
```

```python
n_samples = X_train.shape[0]
n_feats = X_train.shape[1]
n_classes = y_train.shape[1]
```

```python
# convert y vector to categorical
y_train = to_categorical(y_train, num_classes=n_classes)
y_val = to_categorical(y_val, num_classes=n_classes)
```

```python
#X_train = X_train.values.reshape(n_samples, n_feats, 1, 1)
#y_train = y_train.reshape(n_samples, n_classes, 1, 1)
```

```python
print('training shape: ', X_train.shape)
print('training labels: ', y_train.shape)
```

```python
# simple cnn network

model = Sequential()
model.add(Conv2D(32, kernel_size=(2, 1), activation='relu', input_shape=(n_feats,)))
#model.add(MaxPooling2D(pool_size=(2, 1)))
#model.add(Flatten()) 
model.add(Conv2D(64, (2, 1), activation='relu'))
model.add(Dropout(0.5)) 
model.add(Dense(n_classes, activation='softmax'))
model.compile(loss=categorical_crossentropy, optimizer=Adam(), metrics=['accuracy'])
#model.fit(X_train, y_train, batch_size=128, epochs=12, verbose=1, validation_data=(X_val, y_val))
```

```python
model = Sequential()
model.add(Reshape((9, 1), input_shape=(n_feats,)))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(100, activation = 'relu'))
model.add(Flatten())
model.add(Dense(n_classes, activation='softmax'))
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=128, verbose=1, validation_data=(X_val, y_val))
```
